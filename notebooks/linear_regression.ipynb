```python
# notebooks/linear_regression.ipynb (can also run as .py)
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# --- Synthetic data: y = 3x + 2 + noise
rng = np.random.default_rng(42)
X = rng.uniform(-3, 3, size=(200, 1))
y = 3 * X[:, 0] + 2 + rng.normal(0, 0.7, size=200)

# ---------- From scratch (gradient descent) ----------
X_b = np.c_[np.ones((len(X), 1)), X]  # add bias term
theta = np.zeros(2)                   # [bias, weight]
alpha, epochs = 0.05, 1000

for _ in range(epochs):
    preds = X_b @ theta
    grads = (2/len(X_b)) * X_b.T @ (preds - y)
    theta -= alpha * grads

b_hat, w_hat = theta
y_pred_scratch = X_b @ theta
print("[Scratch]  b=%.3f, w=%.3f, MSE=%.3f, R2=%.3f" % (b_hat, w_hat, mean_squared_error(y, y_pred_scratch), r2_score(y, y_pred_scratch)))

# ---------- With scikit-learn ----------
lin = LinearRegression().fit(X, y)
y_pred_sklearn = lin.predict(X)
print("[sklearn]  b=%.3f, w=%.3f, MSE=%.3f, R2=%.3f" % (lin.intercept_, lin.coef_[0], mean_squared_error(y, y_pred_sklearn), r2_score(y, y_pred_sklearn)))
```
